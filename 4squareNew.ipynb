```python
def FourSquareCategorySearch_2(category,name, lat, lng,radius, limit_count,catname,whattouse_):
    
    
    limit=50

    url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, ##
            lng, ##
            radius, 
            limit,
            category)##

    # make the GET request
#     print(radius, limit, category)
    results = requests.get(url).json()["response"]["venues"]

    distance_list=[]

    for v in results:
        distance_list.append([(
        name, #0
        lat,  #1
        lng,   #2
        v['name'], #3
        v["location"]["distance"]) ])#4

    Nom=[]
    ii=0
    distance=[]
    for item in distance_list:#for each item outputted
#         print(ii,item)
        ii+=1
        Nom.append(item[0][3])

        ###########Some changes i.e H C S not really a supermarket
        if category=='52f2ab2ebcbc57f1066b8b46':#supermarket    
            if item[0][3]=='Hung Cheong Stores':
                print('hc stores',item[0][4]*5)
                distance.append(item[0][4]*5)
            else:
                distance.append(item[0][4])
        else:
            distance.append(item[0][4])


    #take the min distance
    try:
        distanceOut=np.min(distance)
        #and count how many
        vCount=(sum(jj < limit_count for jj in distance))
        
    except:
        print(catname,name,distance,'prob nothing found')
        distanceOut=16000
        vCount=0
    if whattouse_==0:
        out=distanceOut
    else:
        out=vCount
#     print(out)
    return out
```

```python
def FourSquareCategorySearch(category,names, latitudes, longitudes,radius, limit_count):
    
    #radius=10000#say 10km max is 100km
    limit=50#max 
#     if limit_count not in locals:
#         limit_count=1000#1km
    
    distance_list=[]
    distanceOut=[]
    i=0
    for name, lat, lng in zip(names, latitudes, longitudes):
#         print(i)
        url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                lat, ##
                lng, ##
                radius, 
                limit,
                category)##
        
        # make the GET request
        results = requests.get(url).json()["response"]["venues"]
        

        distance_list.append([(
            name, 
            lat, 
            lng, 
            v['name'], 
            v['location']['lat'], 
            v['location']['lng'],
            v["location"]["distance"], 
            v["categories"][0]['name']) for v in results])
        
        i=i+1
        
        
    vCat=[]
    vNom=[]
    vLAT=[]
    vLONG=[]
    vCount=[]
    for distanceItem in distance_list:#for each location given
        distance=[]
        Cat=[]
        Nom=[]
        LAT=[]
        LONG=[]
        
        for item in distanceItem:#for each category outputted
            
            Cat.append(item[7])
            Nom.append(item[3])
            
            ###########Some changes i.e H C S not really a supermarket
            if category=='52f2ab2ebcbc57f1066b8b46':#supermarket    
                if item[3]=='Hung Cheong Stores':
                    print('hc stores',item[6]*5)
                    distance.append(item[6]*5)
                else:
                    distance.append(item[6])
            else:
                distance.append(item[6])
                
                
            LAT.append(item[4])
            LONG.append(item[5])
            
        #take the min distance
        minDist=np.min(distance)
        posMin=distance.index(minDist)
        distanceOut.append(minDist)
        #
        vCount.append(sum(jj < limit_count for jj in distance))
        
        #
        vCat.append(Cat[posMin])
        vNom.append(Nom[posMin])
        vLAT.append(LAT[posMin])
        vLONG.append(LONG[posMin])
        

#     print(names,'---',vNom,'---',distanceOut,'---',vCount)

    
#     
    
    nearby_venues  =pd.DataFrame(columns=['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Distance_km',
                  'No. Venues within distance' ,                       
                  'Venue Category'])
    for ii in range(0,i):
        nearby_venues.loc[ii] = [names.iloc[ii], latitudes.iloc[ii], longitudes.iloc[ii], vNom[ii] ,vLAT[ii], vLONG[ii], distanceOut[ii]/1000, vCount[ii],vCat[ii]  ]
    
    
    
    return nearby_venues



    ###If want a df of values uncomment below
#     venues_list.append([(
#             name, 
#             lat, 
#             lng, 
#             v['name'], 
#             v['location']['lat'], 
#             v['location']['lng'],
#             v["location"]["distance"], 
#             v["categories"][0]['name']) for v in results])
#     nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
#     nearby_venues.columns = ['Neighborhood', 
#                   'Neighborhood Latitude', 
#                   'Neighborhood Longitude', 
#                   'Venue', 
#                   'Venue Latitude', 
#                   'Venue Longitude', 
#                   'Venue Distance', 
#                   'Venue Category']
#     return(nearby_venues)
```

```python
def getNearbyVenuesOLD(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['icon']['prefix'].split('/')[-2]) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
