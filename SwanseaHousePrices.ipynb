## Add some defs here to avoid repeating

```python
def housepriceClean(dfHP):
    #load big file to convert postcodes to census id
    import pandas as pd
    df=pd.read_csv('Swansea_step1.csv',usecols=["Postcode","CensusID"])
    #get rid of other property types
    dfHP=dfHP[dfHP.Type!="Other"]
    #merge so get census ID
    dfHP=pd.merge(dfHP,df,on='Postcode')
    
    #take years between 2008 and 2019 inclusive
    #first convert to years and addas a column
    from datetime import datetime
    datAll=[]
    for dat in dfHP.Date:
        datAll.append(datetime.strptime(dat,'%Y-%m-%d').year)
    #add new column named Year
    dfHP['Year']=datAll
    #just middle years- why won't cond work together with and/&?
    cond=dfHP.Year>2007
    dfHP=dfHP[cond]
    cond=dfHP.Year<2019
    dfHP=dfHP[cond]
    return dfHP
```

# Find percent of area by house type

```python

typeSumm=salesSwanCensus.groupby(["GEOGRAPHYCODE","Type"]).agg({"Count":"sum"})
# typeSumm=typeSumm/typeSumm.sum()
type_pcts = typeSumm.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
type_pcts
```

```python
salesSwanCensus
```

```python
salesSwanCensus=pd.merge(type_pcts,salesSwanCensus,how='left',left_on=['GEOGRAPHYCODE','Type'],right_on=['GEOGRAPHYCODE','Type'])
salesSwanCensus
```

```python
coluse=['GEOGRAPHYCODE', 'PcDetached', 'PcSemiD', 'PcTerraced', 'PcFlatpurposebuilt','PcFlatofHse','PcFlatCommercial', 'Price']
census=pd.read_csv('SwanseaCombo_WivPricesWiv4sq.csv',usecols=coluse)
census['PcFlats']=census[['PcFlatpurposebuilt','PcFlatofHse','PcFlatCommercial']].sum(axis=1)
census.drop(columns=['PcFlatpurposebuilt','PcFlatofHse','PcFlatCommercial'])
census=census[['GEOGRAPHYCODE', 'PcDetached', 'PcSemiD', 'PcTerraced', 'PcFlats', 'Price']]
```

```python
census
```

- go through each district
- House price X Census % 

```python
uniqGEO=salesSwanCensus.GEOGRAPHYCODE.unique()

typeChoi=salesSwanCensus.Type.unique()
e=[]
f=[]
for dista in uniqGEO:
    distaDF=salesSwanCensus[salesSwanCensus.GEOGRAPHYCODE==dista]
    a=distaDF[distaDF.Type==typeChoi[0]].Price.values*census[census.GEOGRAPHYCODE==dista].PcDetached.values/100
    b=distaDF[distaDF.Type==typeChoi[1]].Price.values*census[census.GEOGRAPHYCODE==dista].PcSemiD.values/100
    c=distaDF[distaDF.Type==typeChoi[2]].Price.values*census[census.GEOGRAPHYCODE==dista].PcTerraced.values/100
    d=distaDF[distaDF.Type==typeChoi[3]].Price.values*census[census.GEOGRAPHYCODE==dista].PcFlats.values/100
    if a.size==0:
        a=0
    if b.size==0:
        b=0
    if c.size==0:
        c=0
    if d.size==0:
        d=0
    e.append(a+b+c+d)
    f.append(dista)
```

```python
salesSwanCensus
```

```python
priceNew=pd.DataFrame(e)
priceNew['GEOGRAPHYCODE']=f

priceNew=pd.merge(priceNew,salesSwanCensusAll,left_on='GEOGRAPHYCODE',right_on='GEOGRAPHYCODE')
priceNew=priceNew.rename(columns={'Price':'Old Price',0:"PriceNew"})
priceNew
```

```python
priceNew=priceNew[priceNew.Count>3]
priceNew
```

```python
census=pd.merge(census,priceNew,left_on='GEOGRAPHYCODE',right_on='GEOGRAPHYCODE')
census=census.rename(columns={'Price':"Old Price"})
census=census[['GEOGRAPHYCODE','PriceNew']]
census
```
