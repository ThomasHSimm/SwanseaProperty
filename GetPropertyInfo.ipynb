```python
import pandas as pd
import numpy as np
iwanthatone=['Postcode','In Use?','Latitude','Longitude','Parish','LSOA Code']
df=pd.read_csv('_data/Swansea postcodes.csv',usecols=iwanthatone)
#lets remove some ones we don't need
df=df[df['In Use?']=='Yes']
df=df.drop(columns=['In Use?','LSOA Code'])

# alternative
# df=pd.read_csv('_data/Swansea.csv')
# df=df[df['In Use?']=='Yes']
# df
df
```

```python
#webscrape to get census ID
def get_from_doogal(df):
    newdata=[]#initialise the series
    ii=0
    for pc1 in df.Postcode:
        pc1=pc1.replace(' ','%20')
        url='https://www.doogal.co.uk/ShowMap.php?postcode={}'.format(pc1)
        
        print(url)
        
        aa=pd.read_html(url)

        ## The demographics table
        aaX=aa[4]
        
        #print(aaX)
        
        # Population
        str_check='Population'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)###PAIN!!!! if don't do this difficult to get a string to split!!!!
            Pop=int(aaX[1][boola])
#             print(str_check,Pop)
        else:
            Pop=0
#             print('No pop stuff')#,boola)
            
        # Households
        str_check='Households'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)###PAIN!!!! if don't do this difficult to get a string to split!!!!
            house=int(aaX[1][boola])
#             print(str_check,house)
        else:
            house=0
#             print('No house stuff')#,boola)
        
        # Income
        str_check='income'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)
            income=(aaX[1][boola])
            #print(income)
            income=int(str(income).split('L')[0].split('Â£')[1].replace(',',''))
#             print(str_check,income)
        else:
            income=0
#             print('No income stuff')
            
        # Census
        str_check='Census'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)
            censusNo=(aaX[1][boola])
            
#             print(str_check,censusNo)
        else:
            censusNo=0
#             print('No census stuff')
        

        # Middle layer area id
        str_check='Middle layer'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)
            district=str(aaX[1][boola])
#             print(str_check,district)
        else:
            district=""
#             print('No district stuff')

        # DEprivation
        str_check='Deprivation'
        boola=aaX[0].str.contains(str_check)
        
        if boola.any():
            boola=int(boola[boola].index.values)
            depr=str(aaX[1][boola])
            depr=depr.split('/')#[0].split('}')#[-1].replace(',','')))
            depr=int(depr[0].split('}')[-1].replace(',',''))
#             print(str_check,depr)
        else:
            depr=0
#             print('No dep stuff',boola)
        
        ## Add data to series->>>> #better / quicker not to append a DF    
        newdata.append([district,income, depr, Pop,house,censusNo])
        print(ii)
        ii=ii+1
    #print(newdata)
    return newdata
```
